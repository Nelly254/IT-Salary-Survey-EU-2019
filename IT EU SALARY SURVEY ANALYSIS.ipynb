# <center>1. Importing Libraries</center>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

# <center>2. Loading The Dataset</center>

```python
df = pd.read_csv('T Salary Survey EU 2019.csv')
df
```

```python
df.info()
```

```python
# Dimension of the Data
df.shape # 991 rows and 23 columns
```

We plot a heatmap of the null values.

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.isna());
```

Checking for NaN value

```python
df.isna().sum()
```

###### Data Cleaning for Gender Column

```python
feature = 'Gender'
df[feature].unique()
```

```python
df[[feature]] = df[[feature]].fillna(value='Not Specified')
df[feature].unique()
```

```python
df['Age'].unique()
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(x='Gender',y='Age',data=df,palette='Pastel2')
plt.xticks(fontsize=13)
plt.xlabel("")
plt.yticks(fontsize=13)
plt.ylabel("Age",fontsize=14)
plt.show()
```

```python
salary = 'Yearly brutto salary (without bonus and stocks)'

plt.figure(figsize=(15,8)) 
ax = sns.boxplot(x='Gender',y=salary,data=df,palette='Set2',linewidth=2.5)
ax = sns.stripplot(x='Gender',y=salary,data=df)

plt.xticks(fontsize=13)
plt.xlabel("")
plt.yticks(fontsize=13)
plt.ylabel(salary,fontsize=14)
plt.show()
```

Distribution of Yearly Salaries

```python
plt.figure(figsize=(15,8))
sns.histplot(x='Yearly brutto salary (without bonus and stocks)',data=df,
             bins=20, kde=True, hue='Gender',multiple="stack",palette='magma')
plt.xticks(fontsize=13)
plt.xlabel("Yearly brutto salary (without bonus and stocks)",fontsize=14)
plt.yticks(fontsize=13)
plt.ylabel("Count",fontsize=14)
plt.show()
```

Fewer females are noted in this sector, compared to the males.
